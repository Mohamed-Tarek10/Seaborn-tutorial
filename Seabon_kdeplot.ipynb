#### A dataset that is built-in that come with the library

```python
sns.relplot(x='model_year',    # X_axis 
              y= 'mpg',        # Y_axis
            col = 'origin',    # Creates a subplot (column) based on unique values. Each column represents a distinct value(usa, japan, europe)
            hue = 'cylinders', # Colors dataframe according to the number of cylinders, helping distinguish engine sizes 
            data =cars);       # Specifies the data 'cars' as a data source
```

#### Univariate KDEplot (1D)

```python
sns.kdeplot(cars.horsepower)
sns.set_style('darkgrid')
```

```python
sns.kdeplot(cars.horsepower,
            shade = True  # Shadow
           )
```

```python
sns.kdeplot(cars.horsepower,
            bw = 5,    # BandWidth
           shade = True,
           )
```

```python
sns.kdeplot(cars.horsepower,
            cumulative = True   # If we want to switch a probability density function with cumulative function
           )
```

#### Bivariate KDEplot (2D)

```python
sns.kdeplot(x = cars.horsepower, y= cars.mpg)
```

```python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, n_levels = 30  # Increase the number of granularity
           )
```

```python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, n_levels = 30,  # Increase the number of granularity
           bw_adjust = .5)
```

```python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, n_levels = 30,  # Increase the number of granularity
           bw_adjust = .5, # Control the bandwidth (smoothing)
            fill=True      # Make the figure look smoother with colour filling
           )
```

```python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, shade =True)
```

```python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, shade = False, cbar = True)
```

```python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, shade = True, cbar = True)
```
