```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
"%matplotlib inline"
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
diamonds.duplicated( keep = False).sum()
```

```python
diamonds.drop_duplicates(inplace = True)
```

```python
'''
We notice that the outout in the first line is boolean values. Then, We do summtion for this values to
get the columns that contain ( NaN ) Values. 

When we sum "False" + "False" == 0 + 0 == 0
When we sum "False" + "True" == 0 + 1 == 1
When we sum "True" + "True" == 1 + 1 == 1
'''
d = diamonds.isna()      
d.sum()
```

```python
pd.DataFrame(diamonds[['depth', 'table', 'price', 'x', 'y', 'z']].value_counts(dropna = False))
```

```python
# We notice above tat thee is a row that have many NaN values in it. So we can drop it.
diamonds.dropna(axis = 0)
```

```python
sns.scatterplot(x = diamonds.carat, y = diamonds.price ,hue = 'color', data = diamonds,
                #legend = None  # We can put this to delete the legend box.
               )
plt.legend(bbox_to_anchor = (1,1))
```

```python
sns.scatterplot(x = diamonds.carat, y = diamonds.price ,hue = 'cut', data = diamonds,
                palette = ['purple', '#55cccc']  # We can also use different colors.
               )
plt.legend(bbox_to_anchor = (1,1))
```

```python
sns.scatterplot(x = diamonds.carat, y = diamonds.price ,hue = 'color', data = diamonds,
                #legend = None  # We can put this to delete the legend box.
                 s = 100,        # Give us the size of the scatter balls.
                 style = 'cut'   # Create a new look for the points.
               
               )
plt.legend(bbox_to_anchor = (1,1))
```

```python
# Using " Alpha "
# A lot of points is overlapping wih each other, so we can use "alpha" to avoid that.
# "Alpha" is a range from (0:1)
# '0' refers to invisible
# '1' refers to completely solid

sns.scatterplot(x = diamonds['carat'], y = diamonds['price'], hue = 'clarity', alpha = 0.9, marker = '*',data = diamonds)

# We also can change the marker style, by using "marker" argument = " * "
```
